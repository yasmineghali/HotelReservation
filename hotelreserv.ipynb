Descion Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X = df.drop(['booking_status','Booking_ID'],axis=1)
y = df.booking_status
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train,y_train)
print("Accuracy : ", tree_clf.score(X_train,y_train)*100)
print("Accuracy : ", tree_clf.score(X_test,y_test)*100)
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
logModel = LogisticRegression()
logModel.fit(X_train,y_train)
Y_pred = logModel.predict(X_test)
print("Accuracy : ", logModel.score(X_train,y_train)*100)
```

```python
print("Accuracy : ", logModel.score(X_test,y_test)*100)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
l=[]
for n_neighbors in range(1,50):
 knn = KNeighborsClassifier(n_neighbors)
 knn.fit(X_train, y_train)
 y_pred = knn.predict(X_test)
 l.append(metrics.accuracy_score(y_test,y_pred))
```
